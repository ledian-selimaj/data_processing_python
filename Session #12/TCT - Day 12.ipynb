### random numbers

```python
from numpy import random as rand
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def plot_distribution(x, axs=None, xlim=None, ylim=None, figsize_=(15,8), bins_=100, kde_=True, cumulative_=False, xlabel='Distribution'):
    # plot the data
    if axs == None:
        fig, axs = plt.subplots(figsize=figsize_)
    axs.hist(x, density=False, bins=bins_, facecolor='b', alpha=0.5)
    axs.grid(alpha=0.75)
    axs.set_xlabel(xlabel)
    axs.set_ylabel('Count')
    # axs.set_title('Histogram plot of a random variable')

    # axs.set_xlim([x.min(), x.max()])
    if xlim:
        x_min, x_max = xlim
    else:
        x_min, x_max = axs.get_xlim()
    if ylim:
        y_min, y_max = ylim
    else:
        y_min, y_max = axs.get_ylim()
    axs.set_xlim([x_min, x_max])
    axs.set_ylim([y_min, y_max])
    # axs.yaxis.set_major_formatter(matplotlib.ticker.FormatStrFormatter('%.3e'))
```

```python
triangular=rand.triangular(29,150,200,1000000)
```

```python
triangular.shape
```

```python
triangular[0]
```

```python
triangular[0:10]
```

```python
plot_distribution(triangular)
```

```python
uniform=rand.uniform(50,100,100000)
```

```python
plot_distribution(uniform)
```

```python
gamma=rand.gamma(2,4,1000000)
```

```python
plot_distribution(gamma)
```

### Visulization using seaborn

```python
import seaborn as sns
```

```python
sns.set(color_codes=True)
```

```python
auto=pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
sns.set({'figure.figsize':(8,6)})
sns.histplot(auto['highway_mpg'], kde=True)
```

```python
#sns.set({'figure.figsize':(8,6)})
sns.histplot(auto['engine_size'], kde=True)
```

```python
#sns.set({'figure.figsize':(8,6)})
sns.jointplot(x='engine_size',y='horsepower',data=auto)
```

```python
sns.jointplot(x='engine_size',y='horsepower',data=auto,kind='hex')
```

```python
sns.jointplot(x='engine_size',y='horsepower',data=auto,kind='kde')
```

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
```

```python
sns.boxplot(auto['horsepower'],orient='h')
```

```python
sns.boxplot(x='number_of_doors',y='horsepower', data=auto,orient='v')
```

```python
sns.boxplot(x='number_of_doors',y='horsepower', data=auto, hue=auto['fuel_type'],orient='v')
```

```python
sns.barplot(x='body_style', y='horsepower', data=auto,  hue=auto['fuel_type'] )
```

```python
sns.countplot(x='body_style', data=auto, hue=auto['fuel_type'])
```

```python
sns.catplot(x='fuel_type',
           y='horsepower',
           col='drive_wheels',
           hue='number_of_doors',
           data=auto,
           kind='box')
```

```python
sns.lmplot(x='engine_size', y='horsepower', data=auto )
```

```python
sns.lmplot(x='engine_size', y='horsepower', hue='fuel_type', data=auto )
```
