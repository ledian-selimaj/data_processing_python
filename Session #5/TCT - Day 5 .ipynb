### Combining pandas dataframes

#### concat
(we will build two dataframes from dictionaries for this purpose)

```python
dict1 = {
  "name": ["Eni","Flavia", "Andi", "Frensi", "E"] ,
  "math": [2,4,5,7,8],
  "chemisrty": [7,8,9,10,8],
  "physics":[3,6,7,9,8]
}
```

```python
dict1
```

```python
dict2 = {
  "name": ["Andi","Eni","C","D","E"], 
  "math": [7,8,9,5,8],
  "chemisrty": [6,8,9,4,8],
  "physics":[10,5,8,6,8]
}
```

```python
dict2
```

```python
#convert dictionaries to dataframes
df1=pd.DataFrame(dict1)
df2=pd.DataFrame(dict2)
```

```python
print(df1.head())
print(df2.head())
```

```python
#using concat, default operation is adding rows in one df, similar to union in SQL
pd.concat([df1,df2])
```

```python
#reset indexes for all concatenated rows
pd.concat([df1,df2]).reset_index(drop=True)
```

```python
#reset indexes for all concatenated rows
pd.concat([df1,df2],ignore_index=True)
```

```python
#reset indexes for all concatenated rows and drop duplicates 
df3=pd.concat([df1,df2]).drop_duplicates().reset_index(drop=True)
df3
```

```python
#concat on column using axis parameter
pd.concat([df1,df2],axis=1).drop_duplicates().reset_index(drop=True)
```

```python
df3.info()
```

```python
#create a new columns that is a concatenation of two columns. They should be
#same type. We need to converet from int to object in our case
df3['concat'] = df3['name'] +'-' + df3['math'].astype(str)
df3.info()
```

```python
#applying substring in string column (last letter of column name), create new column 
df3['substring'] = df3['name'].str.slice(-1)
df3.head()
```

```python
#example of indexing single string value
name = 'Ledian'
name[-1]
```

#### merge
(important parameters, how and on)

```python
#performing left join on name, same column names take _x and _y endings
#missing values show as NaN
pd.merge(df1,df2,how='left', on = 'name')
```

```python
#performing right join on name, same column names take _x and _y endings
#missing values show as NaN
pd.merge(df1,df2,how='right', on = 'name')
```

```python
#performing outer (full) join on name, same column names take _x and _y endings
#missing values show as NaN
pd.merge(df1,df2,how='outer', on = 'name')
```

```python
#fill Nan with 0
pd.merge(df1,df2,how='outer', on = 'name').fillna(0)
```

```python
#performing inner join on name, same column names take _x and _y endings
pd.merge(df1,df2,how='inner', on = ['name'])
```

```python
#equivalent operation on merge
df1.merge(df2,how='inner', on = ['name'])
```

```python
#rename key column to then merge
df2.rename(columns={'name':'name1'},inplace=True)
```

```python
df2.head()
```

```python
#using left_on and right_on in case the keys have different names, drop one of the key  columns
pd.merge(df1,df2,how='inner', left_on = 'name',right_on = 'name1').fillna(0).drop(columns={'name1'})
```

#### join
(similar to merge, but join happen on the index. We normally use merge and join only in special cases when we need to join row from one df with row on another df on same index values)

```python
df1.head()
```

```python
df2.head()
```

```python
#in our example, tables have same names, so we need to define a suffixfor left and right tables
#this example is similar to concat on column level using axis=1 
df1.join(df2,how='inner',lsuffix='_left',rsuffix='_right')
```

### datetime operations

```python
dict3={
    'date':['15/03/2001','15/03/2003','03/03/2005','03/03/2010'],
    'math':[7,8,9,10],
    'chemistry':[1,2,7,1],
    'physics':[2,3,4,6]
}
```

```python
df4=pd.DataFrame(dict3)
```

```python
df4
```

```python
df4.info()
```

```python
#convert an object type to datetime in pandas
df4['date_dt']=pd.to_datetime(df4['date'],dayfirst=True)
```

```python
df4.info()
```

```python
df4
```

```python
#extract year, month, day from datetime
df4['year']=df4['date_dt'].dt.year
df4['month']=df4['date_dt'].dt.month
df4['day']=df4['date_dt'].dt.day
```

```python
df4
```

```python
#add new column
df4['date_dt_2']=['15/03/2011','10/04/2007','01/08/2001','02/03/2020']
```

```python
#convert to datetime
df4['date_dt_2']=pd.to_datetime(df4['date_dt_2'])
```

```python
#get date difference between two columns in months
df4['date_diff_month']=df4['date_dt_2'].dt.month-df4['date_dt'].dt.month
```
