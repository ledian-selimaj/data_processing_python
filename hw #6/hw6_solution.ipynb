```python
#Q1: Convert ages to groups of age ranges: ≤12, Teen (≤ 18), Adult (≤ 60) and Older (>60)
#hint: use the function cut(), create a function and pass it as parameter in a pipe function
#hint: use function assign to create a new column in the chain
```

```python
bins=[0, 13, 19, 61, 100]
labels=['<12', 'Teen', 'Adult', 'Older']
res=(
    data
    .assign(GroupAge = lambda df: pd.cut(df['Age'],bins=bins,labels=labels))
)
res.head()
```

```python
def create_age_group(x_df):
    bins=[0, 13, 19, 61, 100]
    labels=['<12', 'Teen', 'Adult', 'Older']
    ageGroup= pd.cut(x_df['Age'],bins=bins,labels=labels)
    x_df['ageGroup']=ageGroup
    return x_df
```

```python
data['Age']= data['Age'].astype(float)
```

```python
data.info()
```

```python
res=(
    data
    .pipe(create_age_group)
)
res.head()
```

```python
#Q2: Select passengers embarked from Southampton (S)
#hint: use query function
```

```python
bins=[0, 13, 19, 61, 100]
labels=['<12', 'Teen', 'Adult', 'Older']
res=(
    data
    .assign(GroupAge = lambda df: pd.cut(df['Age'],bins=bins,labels=labels))
    #.pipe(create_age_group)
    .query('Embarked == "S"')
)
res.head()
```

```python
#Q3: Create a pivot table to display the survival rate for different age groups and Pclass
#hint: call pivot function in the chain along with the necessary parameters
```

```python
bins=[0, 13, 19, 61, 100]
labels=['<12', 'Teen', 'Adult', 'Older']
res=(
    data
    .assign(GroupAge = lambda df: pd.cut(df['Age'],bins=bins,labels=labels))
    #.pipe(create_age_group)
    .query('Embarked == "S"')
    .pivot_table(values = 'Survived', index='Pclass', columns = 'GroupAge', aggfunc='mean') 
)
res
```

```python
#Q4: Improve the display of pivot table by renaming axis labels and formatting values.
```

```python
bins=[0, 13, 19, 61, 100]
labels=['<12', 'Teen', 'Adult', 'Older']
res=(
    data
    .assign(GroupAge = lambda df: pd.cut(df['Age'],bins=bins,labels=labels))
    #.pipe(create_age_group)
    .query('Embarked == "S"')
    .pivot_table(values = 'Survived', index='Pclass', columns = 'GroupAge', aggfunc='mean') 
    .rename_axis('',axis="columns")
    .style.format('{:.2%}')
)
res
```
