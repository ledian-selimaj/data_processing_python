### iteritems

```python
for i, row in data.iteritems():
    print('Column: ',i)
    print(row)
```

```python
columns=list(data.columns)
```

```python
for column in columns:
    print(data[column])
```

```python
for column in columns:
    #print(column)
    if column=='Open':
        print(data[column][2])
```
