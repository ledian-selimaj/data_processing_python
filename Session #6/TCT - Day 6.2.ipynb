```python
import pandas as pd
```

```python
df = pd.read_csv('pivot_example.csv')
df
```

```python
df.pivot_table(index='Name')
```

```python
df.Name.unique()
```

```python
df[df['Name']=='Alisa'].mean()
```

```python
df.pivot_table(index='Name', aggfunc=['sum','count'])
```

```python
df.pivot_table(index='Name', aggfunc=['sum','count'], values='Score')
```

```python
df.pivot_table(index='Name', aggfunc=['sum','count'], values='Score', columns=['Subject'])
```

```python
df.pivot_table(index='Name', aggfunc=['sum','count'], values='Score', columns=['Subject', 'Exam'])
```

```python
df.pivot_table(index=['Name','Subject', 'Exam'],aggfunc=['sum','count'], values='Score')
```
